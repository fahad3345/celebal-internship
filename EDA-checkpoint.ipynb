```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### EDA on titanic dataset

```python
titanic=pd.read_csv('titanic.csv')
titanic
```

```python
#first 5 data
titanic.head()
```

```python
#last 5 data
titanic.tail()
```

```python
#describe mean, count, standered devaition, 25,50,75 percentile and max value
titanic.describe()
```

```python
#checking is there any NaN values is in dataset
titanic.isna().sum()
```

```python
#total survived
titanic['Survived'].count()
```

```python
#bar plot of count of survived
plt.figure(figsize=(4,3))
sns.countplot(data=titanic,x='Survived')
plt.show()
```

```python
#count of survived people with their sex
titanic.groupby(['Sex', 'Survived'])['Survived'].count()
```

```python
plt.figure(figsize=(4,3))
sns.countplot(data=titanic,x='Sex',hue='Survived')
plt.show()
```

```python
# male survived or not survived using pie chart
titanic[titanic['Sex'] == 'male'].Survived.groupby(titanic.Survived).count().plot(kind='pie', figsize=(4, 6),explode=[0,0.05],autopct='%1.1f%%')
plt.axis('equal')
plt.legend(["Not Survived","Survived"])
plt.title("Male survival rate")
plt.show()
```

```python
# female survived or not survived using pie chart
titanic[titanic['Sex'] == 'female'].Survived.groupby(titanic.Survived).count().plot(kind='pie', figsize=(4, 6),explode=[0,0.05],autopct='%1.1f%%')
plt.axis('equal')
plt.legend(["Not Survived","Survived"])
plt.title("Female survival rate")
plt.show()
```

```python
#count of survived people with their class
sns.countplot( data=titanic,x='Pclass', hue='Survived')
plt.title('Pclass: Sruvived vs Dead')
plt.show()
```

```python
pd.crosstab(titanic.Pclass, titanic.Survived, margins=True)
```

```python
sns.barplot(data=titanic,x='Pclass',y='Survived',hue='Sex')
```

```python
#count of survived with embarked(embarked means from where like Southampton, Cherbourg, and Queenstown)
sns.countplot(x='Survived', data=titanic,hue = 'Embarked');
```

```python
#fare with each class
sns.barplot(y = "Fare",x = "Pclass",data = titanic)
```

### EDA on house price prediction

```python
house_price=pd.read_csv('boston_house_prices.csv',skiprows=1)
house_price.rename(columns={'MEDV':'PRICE'},inplace=True)
```

```python
house_price
```

```python
#first 5 data
house_price.head()
```

```python
#last 5 data
house_price.tail()
```

```python
#describe mean, count, standered devaition, 25,50,75 percentile and max value
house_price.describe()
```

```python
#checking is there any NaN values is in dataset
house_price.isna().sum()
```

```python
#columns with types and non-null count
house_price.info()
```

```python
#Univariate Analysis
# Plotting all the columns to look at their distributions
for i in house_price.columns:
    plt.figure(figsize = (8, 3))
    sns.histplot(data = house_price, x = i, kde = True,color='blue')
    plt.show()
```

```python
#Check the correlation using heatmap
plt.figure(figsize = (12, 8))
cmap = sns.diverging_palette(230, 20, as_cmap = True)
sns.heatmap(house_price.corr(), annot = True, fmt = '.2f', cmap = cmap)
plt.show()
```

### EDA on Churn Dataset

```python
churn=pd.read_csv('Churn_Modelling.csv')
churn
```

```python
churn.shape
```

```python
#first 5 data
churn.head()
```

```python
#last 5 data
churn.tail()
```

```python
#describe mean, count, standered devaition, 25,50,75 percentile and max value
churn.describe()
```

```python
#checking missing values
churn.isna().sum()
```

```python
#checking of type of each column
churn.info()
```

```python
#Data wrangling
churn.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1, inplace=True)
```

```python
churn.head()
```

```python
#count of geography
sns.countplot(x=churn['Geography'])
```

```python
#count of gender
sns.countplot(x=churn['Gender'])
```

```python
#check the correlation between each
plt.figure(figsize=(10,10))
# Select only the numeric columns
numeric_columns = churn.select_dtypes(include=[np.number])

# Calculate the correlation between numeric columns
correlation = numeric_columns.corr()

# Plot the correlation heatmap
sns.heatmap(data=correlation,annot=True,fmt='0.1g')
plt.show()
```

```python
#Let's check the distribution of Credit Score first.
plt.figure(figsize=(15, 8))
sns.histplot(churn.CreditScore,kde=True,color='blue')
plt.ylabel('Frequency', fontsize=15)
plt.xlabel('Credit Score', fontsize=15)
plt.title('Credit Score Frequency Distribution', fontsize=15)
plt.show()
```

```python
#Let's check the relation of credit score with geography.
list(churn.Geography.unique())
```

```python
plt.figure(figsize=(15, 8))
sns.catplot(kind='box', x='CreditScore', data=churn, col='Geography')
```

```python
#let's check the relation of credit score with gender
plt.figure(figsize=(15, 8))
sns.catplot(kind='box', x='CreditScore', data=churn, col='Gender')
```

```python
#let's check credit score with age
plt.figure(figsize=(15, 8))
sns.scatterplot(x='CreditScore', data=churn, y='Age')
```

```python
#count of Geographical locations
plt.figure(figsize=(15, 8))
sns.countplot(x=churn.Geography)
plt.xlabel('Geography', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.title('Count of Geographical Locations', fontsize=15)

plt.show()
```

```python
#Let's first try to see if there is any relation of Geography with Tenure. Perhaps, different countries have different tenures.
churn[['Tenure', 'Geography']].dtypes
```

```python
#Let's look at the unique years of tenure first to get the idea.
print('Unique years of tenures:', list(np.sort(churn.Tenure.unique())))
```

```python
#count of each tenure
plt.figure(figsize=(15, 8))
sns.countplot(x=churn['Tenure'])
plt.title('Count of each tenure', fontsize=15)
plt.show()
```

```python
#count plot of gender
sns.countplot(x=churn.Gender)
```

```python
#count of gender with Exited or not
sns.countplot(data=churn,x='Gender',hue='Exited')
```

```python
#find unique values of column age
np.sort(churn.Age.unique())
```

```python
#Let's check the age bracket of those churned and those not churned.
plt.figure(figsize=(15,8))
sns.boxplot(data=churn,x='Age',y='Exited',orient='h')
plt.show()
```

```python
#head of balance amount
churn.Balance.head(10)
```

```python
#check balance amount with exited or not
sns.boxplot(data=churn,x='Balance',y='Exited',orient='h')
```

```python
#check the correlation between Balance amount and Exited
churn[['Balance', 'Exited']].corr()
```
